```python
emissions_csv = pd.read_csv('input_data/emissions.csv')
clean_emissions = emissions_csv.loc[emissions_csv["State"].isin(['OH', 'PA', 'TX', 'WA', 'VT'])]
clean_emissions
```

```python
# A bar chart just looked so sloppy
# I decided to go with a heat map and use population as the "weight"

# These are my x and y values
y = [new_df['Population']]
x = np.arange(len(new_df))
tick_locations = [value for value in x]

plt.figure(figsize=(50,8))
plt.bar(x, y, color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, x["State"], rotation="vertical")



# population = states['Population'].astype(float)
# plt.bar(states['State'], population)
# plt.set_xticklabels(states['State'], rotation=90, horizontalalignment = 'right', fontsize = '12')
# plt.set_title('Population By State (2018)', fontsize = '20')
# plt.set_ylabel('Population per State (Million)', fontsize = '12')
# plt.show()


# # plt.bar(x_axis, data, color='g', alpha=0.75)
# # plt.ylim(0, 40000000, 1)
# # tick_locations = [value for value in x_axis]
# # plt.xticks(tick_locations, rotation = 90, ha="right")
# # plt.show
```
