```python
## import dependencies
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
## Import csv files
production_csv = pd.read_csv("input_data/SolarProduction.csv")
emissions_csv = pd.read_csv("input_data/emissions.csv")
```

# Data Munging

```python
production_df = pd.DataFrame (production_csv)

## Need production table to have the same columns as emissions
production_df.drop(["2018","Grand Total"], axis=1, inplace=True) 
production_df= production_df.rename(columns={"STATE": "State"})

## Production data is not available for the states AK, DC, ND, NJ, and WV
## Need to drop the rows'US-TOTAL' and 'Grand Total'
clean_df = production_df.loc[production_df["State"] != "US-TOTAL"]
clean_prod_df = clean_df.loc[clean_df["State"] != "Grand Total"]

clean_prod_df.head()
```

```python
emissions_df = pd.DataFrame (emissions_csv)

## Need to drop the states AK, DC, ND, NJ, and WV
drop_AK_df = emissions_df.loc[emissions_df["State"] != "AK"]
drop_DC_df = drop_AK_df.loc[drop_AK_df["State"] != "DC"]
drop_ND_df = drop_DC_df.loc[drop_DC_df["State"] != "ND"]
drop_NJ_df = drop_ND_df.loc[drop_ND_df["State"] != "NJ"]
clean_emis_df = drop_NJ_df.loc[drop_NJ_df["State"] != "WV"]

clean_emis_df.head()
```

```python
## Check that length of the dataframes are the same length
print(len(clean_prod_df))
print(len(clean_emis_df))
```
