```python
station_list=[]
for i in range(len(find_one)):
    station_list.append(find_one[i]['city'])
state_list = []
for i in range(len(find_one)):
    state_list.append(find_one[i]['state'])
lat_list=[]
for i in range(len(find_one)):
    lat_list.append(find_one[i]['lat'])
lon_list=[]
for i in range(len(find_one)):
    lon_list.append(find_one[i]['lon'])
co_df = pd.DataFrame({
    "City":station_list,
    "State":state_list,
    "Latitude":lat_list,
    "Longitude":lon_list
})
co_df
```
